```python
import json
import csv
import requests
import pandas as pd
from config import api_key
```

```python
results_object = []
for x in range(359):
    url ="https://api.data.gov/ed/collegescorecard/v1/schools.json?&_fields=id,school.name,latest.admissions.sat_scores.average.overall,latest.student.FAFSA_applications,latest.admissions.admission_rate.overall&api_key=" + api_key + "&_page=" + str(x)
    #print(requests.get(url))
    #print(requests.get(url).json())
    response = requests.get(url).json()
    #print(json.dumps(response, indent=4, sort_keys=True))
    results_object = results_object + response["results"]
#table3 = pd.DataFrame(response['results'])
#table3
print(results_object)
```

```python
score_table = pd.DataFrame(results_object)
score_table.head()
```

```python
score_table.to_csv("ScoreData.csv")
```
